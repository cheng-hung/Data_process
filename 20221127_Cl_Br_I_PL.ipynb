# Reading Cl exchange data

```python
## Find the subfolders in the current working folder (cwd)
#cwd = os.getcwd()
# cwd = '/nsls2/users/clin1/Documents/Cheng-Hung/temp/20221108_insitu'
cwd = '/Users/chenghunglin/Library/CloudStorage/OneDrive-BrookhavenNationalLaboratory/data_processing/20221121_CsPbBr_growth'
subfolder_Cl = glob.glob(cwd + '/CsPbBr_12mM_Cl/**')
subfolder_Cl.sort()
subfolder_Cl.reverse()
# del subfolder[-1]
print(*subfolder_Cl, sep = '\n')  ## Print one element per line
```

```python
## Reading dark and reference
dark = glob.glob(cwd + '/dark_ref/' + 'Dark**.csv')
ref = glob.glob(cwd + '/dark_ref/' + 'Tol_OA**.csv')

dark_df = pd.read_csv(dark[0], skiprows=5, sep=',')
ref_df = pd.read_csv(ref[0], skiprows=5, sep=',')
```

```python
## Read Cl data into pandas dataframe

df_Cl = pd.DataFrame()
df_norm_Cl = pd.DataFrame()
df_norm2_Cl = pd.DataFrame()

## Wavelength range
w1 = 245
w2 = 700

for i in range(len(subfolder_Cl)):
    data_files = glob.glob(subfolder_Cl[i] + '/**PL**.csv')
    data_files.sort()
    del data_files[0]
    
    ## Data file range
    k1 = 0
    k2 = len(data_files)
    
    df_i = pd.DataFrame()
    skiprows=10
    
    for j in range(k1, k2):
        # df_j = pd.read_csv(data_files[j], names=['wavelength', 'intensity'], skiprows=14, sep='\t')
        df_j = pd.read_csv(data_files[j], skiprows=skiprows, sep=',')
        PL = df_j['Sample']-dark_df.Sample
        df_i = pd.concat([df_i, PL], axis=1, ignore_index=True)
    
    df_Cl[f'{os.path.basename(subfolder_Cl[i])}']= df_i.mean(axis=1)
    #df_norm[f'{os.path.basename(subfolder_Cl[i])}']= df_i.mean(axis=1) / df_i.mean(axis=1).sum(axis=0)
    df_norm_Cl[f'{os.path.basename(subfolder_Cl[i])}']= df_Cl[f'{os.path.basename(subfolder_Cl[i])}'].loc[w1:w2] / df_Cl[f'{os.path.basename(subfolder_Cl[i])}'].loc[w1:w2].sum(axis=0)
    
    
    x = df_Cl[f'{os.path.basename(subfolder_Cl[i])}'].loc[w1:w2].values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
    df_norm2_Cl[f'{os.path.basename(subfolder_Cl[i])}']= x_scaled.flatten()

    
df_Cl['Energy'] = df_j.Energy
df_norm_Cl['Energy'] = df_j.Energy.loc[w1:w2]
#df_norm2['wavelength'] = df_j.wavelength.loc[w1:w2]
df_norm2_Cl.loc[0:w2-w1, 'Energy'] = df_j.Energy.loc[w1:w2].values
```

# Reading I exchange data

```python
## Find the subfolders in the current working folder (cwd)
#cwd = os.getcwd()
cwd = '/Users/chenghunglin/Library/CloudStorage/OneDrive-BrookhavenNationalLaboratory/data_processing/UVVIS_OceanView/20220630_CHLin'
subfolder_I = glob.glob(cwd + '/CsPbBr**')
subfolder_I.sort()
print(*subfolder_I, sep = '\n')  ## Print one element per line
```

```python
## Read I data into pandas dataframe

df_I = pd.DataFrame()
df_norm_I = pd.DataFrame()
df_norm2_I = pd.DataFrame()

## Wavelength range
# w1 = 250
# w2 = 750

for i in range(len(subfolder_I)):
# for i in range(1):
    data_files = glob.glob(subfolder_I[i] + '/**Subt**.txt')
    data_files.sort()
    k1 = 0
    k2 = len(data_files)

    df_i = pd.DataFrame()
    for j in range(k1, k2):
        df_j = pd.read_csv(data_files[j], names=['wavelength', 'intensity'], skiprows=14, sep='\t')

        df_i = pd.concat([df_i, df_j.intensity], axis=1, ignore_index=True)
    
    df_I[f'{os.path.basename(subfolder_I[i])}']= df_i.mean(axis=1)
    #df_norm[f'{os.path.basename(subfolder_I[i])}']= df_i.mean(axis=1) / df_i.mean(axis=1).sum(axis=0)
    df_norm_I[f'{os.path.basename(subfolder_I[i])}']= df_I[f'{os.path.basename(subfolder_I[i])}'].loc[w1:w2] / df_I[f'{os.path.basename(subfolder_I[i])}'].loc[w1:w2].sum(axis=0)
    
    x = df_I[f'{os.path.basename(subfolder_I[i])}'].loc[w1:w2].values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
    df_norm2_I[f'{os.path.basename(subfolder_I[i])}']= x_scaled.flatten()

    
df_I['wavelength'] = df_j.wavelength
df_norm_I['wavelength'] = df_j.wavelength.loc[w1:w2]
#df_norm2['wavelength'] = df_j.wavelength.loc[w1:w2]
df_norm2_I.loc[0:w2-w1, 'wavelength'] = df_j.wavelength.loc[w1:w2].values
```
