```python
# import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import mpld3 
import scipy.signal as scipy
from numpy import diff
from numpy import gradient
plt.rcParams['figure.figsize'] = (4.25, 4.25)
plt.rcParams['font.size'] = '12'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rc('lines', linewidth=1.5)
mpld3.enable_notebook()
%cd /Users/matthewgreenberg/Desktop/MG-NB1-44
%matplotlib notebook
mpld3.enable_notebook()
plt.gcf().subplots_adjust(bottom=0.25)
plt.gcf().subplots_adjust(left=0.25)

#here we are using ASCII data (no heading) output from OceanView software 
#Rhodamine6g is the standard measured here and base file_name in OOceanView
#the output formatting is specified as "file counter" and each consists of a 500 msec scan 


File_name = 'test'

(x,y) = np.loadtxt(File_name+'_Absorbance_'+str(35)+'.txt', delimiter='\t', unpack='true')
(x2,y2) = np.loadtxt(File_name+'_Absorbance_'+str(62)+'.txt', delimiter='\t', unpack='true')
(x3,y3) = np.loadtxt(File_name+'_Absorbance_'+str(108)+'.txt', delimiter='\t', unpack='true')

#subset region of interest of full range of spectrometer
x=x[250:500]
y=y[250:500]
x2=x2[250:500]
y2=y2[250:500]
x3=x3[250:500]
y3=y3[250:500]

#compute first and second derivative by np.gradient
#smoothing used here as noise gets amplified by taking multipe derivatives 

dx=x[1]-x[0]
y=scipy.savgol_filter(y,71,3)
y2=scipy.savgol_filter(y2,71,3)
y3=scipy.savgol_filter(y3,71,3)

dy=np.gradient(y,dx)
dy=scipy.savgol_filter(dy,71,3)
d2y=np.gradient(dy,dx)
dy2=np.gradient(y2,dx)
dy2=scipy.savgol_filter(dy2,71,3)
d2y2=np.gradient(dy2,dx)
dy3=np.gradient(y3,dx)
dy3=scipy.savgol_filter(dy3,71,3)
d2y3=np.gradient(dy3,dx)

d2y=scipy.savgol_filter(d2y,71,3)
d2y2=scipy.savgol_filter(d2y2,71,3)
d2y3=scipy.savgol_filter(d2y3,71,3)

plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Signal')
plt.plot(x,y, 'bo', markersize =.75, label='1 ml/min')
plt.plot(x,d2y/max(d2y)-1, 'b', label='d2y 1ml/min')
plt.plot(x2,y2, 'go', markersize =.75, label='2 ml/min')
plt.plot(x2,d2y2/max(d2y2)-1, 'g', label='d2y 2ml/min')
plt.plot(x3,y3, 'ro', markersize =.75, label='2 ml/min')
plt.plot(x3,d2y3/max(d2y3)-1, 'r', label='d2y 2ml/min')
plt.ylim()
#plt.legend()
a1=np.where(min(d2y)==d2y)
a2=np.where(min(d2y2)==d2y2)
a3=np.where(min(d2y3)==d2y3)

print(x[a1])
print(x[a2])
print(x[a3])
print(1240/x[a1])
print(1240/x[a2])
print(1240/x[a3])
plt.vlines(x[a1], ymin=-1.8, ymax=1.0, colors='b', linestyles='dashed', label='Max Value')
plt.vlines(x[a2],  ymin=-1.8, ymax=1.0, colors='g', linestyles='dashed', label='Max Value')
plt.vlines(x[a3],  ymin=-1.8, ymax=1.0, colors='r', linestyles='dashed', label='Max Value')
```
