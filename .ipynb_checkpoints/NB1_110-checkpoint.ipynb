```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import mpld3 
from scipy import integrate  
import scipy.signal as scipy


from numpy import diff
from numpy import gradient
from scipy.optimize import curve_fit
import seaborn as sns
sns.set_theme()
sns.set(style="ticks", rc={"lines.linewidth": 0.9})
import cycler
n = 20
color = plt.cm.viridis(np.linspace(0, 1,n))
plt.rcParams['axes.prop_cycle'] = cycler.cycler('color', color)
plt.rcParams['figure.figsize'] = (3.25,3.25)
plt.rcParams['font.size'] = '12'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rc('lines', linewidth=1.5)
mpld3.enable_notebook()
%matplotlib notebook
%cd /Users/matthewgreenberg/Documents/NB1_110
mpld3.enable_notebook()
plt.gcf().subplots_adjust(bottom=0.25)
plt.gcf().subplots_adjust(left=0.25)

def gauss(x, H, A, x0, sigma):
    return H + A * np.exp(-(x - x0) ** 2 / (2 * sigma ** 2))
def gauss_fit(x, y):
    mean = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - mean) ** 2) / sum(y))
    popt, pcov = curve_fit(gauss, x, y, p0=[min(y), max(y), mean, sigma])
    return popt

Wavelength = []
fwhm = []
Amplitude = []
IntegratedPl = []
Br = list(range(25,500,25))
Br = [x/1000 for x in Br]
print(Br)

def load_fldatab(x,y):
    for n in range(x,y,20):
        (wave,pl) = np.loadtxt('Fluor_Subt11__'+str(n+20)+'__'+str(n+20)+'.txt', skiprows=14, delimiter='\t', unpack='true')
        plt.xlabel('Wavelength (nm)')
        plt.ylabel('PL Intensity (Counts)')
        wave=wave[300:500]
        pl=pl[300:500]
        plt.plot(wave,pl,label='3:1 Br:CsPb')
        H, A, x0, sigma = gauss_fit(wave,pl)
        Wavelength.append(x0)
        FWHM = 2.355*sigma 
        Integral = integrate.simps(pl,wave)
        IntegratedPl.append(Integral)
        fwhm.append(FWHM)
        Amplitude.append(A)
        
       
        
       
            



load_fldatab(0,180)
load_fldatab(201,400)




print(IntegratedPl)









```

```python
mpld3.enable_notebook()
%matplotlib notebook
plt.plot(Br,fwhm,'o')
plt.xlabel('Br2 Flow Rate')
plt.ylabel(r'$\sigma$')
plt.tight_layout()
```

```python
mpld3.enable_notebook()
%matplotlib notebook
plt.plot(Br,Amplitude,'o')
plt.tight_layout()
plt.xlabel('Br2 Flow Rate')
plt.ylabel(r'Amplitude')
plt.tight_layout()
```

```python
%matplotlib notebook
%cd /Users/matthewgreenberg/Documents/NB1_110

Ab_365 = []
Ab_400 = []

def load_absdatab(x,y):
    for n in range(x,y,20):
        (wave,ab) = np.loadtxt('Abs_Absorbance__'+str(n+20)+'__'+str(n+20)+'.txt', skiprows=14, delimiter='\t', unpack='true')
        plt.xlabel('Wavelength (nm)')
        plt.ylabel('Absorbance')
        Ab_365.append(ab[212])
        Ab_400.append(ab[257])
        wave=wave[200:600]
        ab=ab[200:600]
        plt.plot(wave,ab,'o', markersize=1,)
        plt.tight_layout()


load_absdatab(0,380)
print(Ab_365)
print(Ab_400)
```

```python
mpld3.enable_notebook()
%matplotlib notebook
plt.plot(Br,IntegratedPl,'bo')
plt.tight_layout()
plt.xlabel('Br2 Flow Rate')
plt.ylabel(r'IntegratedPL')
plt.tight_layout()
```

```python
mpld3.enable_notebook()
%matplotlib notebook
RePLQY = [x/y for x,y in zip(IntegratedPl,Ab_365)]
plt.rcParams['figure.figsize'] = (3.25,3.25)
plt.plot(Br,RePLQY/max(RePLQY), 'bo')
plt.tight_layout()
plt.xlabel('Br2 Flow Rate')
plt.ylabel(r'Rel. PLQY')
plt.tight_layout()
```
