## Deconvolute profile with <= 3 peaks

```python
def _multipeak_fit_good_PL(x0, y0, num_peaks, maxfev=100000, fit_boundary=[340, 400, 800], 
                           raw_data=False, plot=False, plot_title=None, dummy_test=False):    
    try:
        w1, _ = find_nearest(x0, fit_boundary[0])
        w2, _ = find_nearest(x0, fit_boundary[1])
        w3, _ = find_nearest(x0, fit_boundary[2])
    except IndexError:
        w1, _ = find_nearest(x0, 340)
        w2, _ = find_nearest(x0, 400)
        w3, _ = find_nearest(x0, 800)
    
    if dummy_test:
        x = x0[w1:w2]
        y = y0[w1:w2]
    else:         
        x = x0[w2:w3]
        y = y0[w2:w3]
    
    mean = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - mean) ** 2) / sum(y))
    
    
    fit_function = [_1gauss, _2gauss, _3gauss]
    initial_guess = [[max(y), mean, sigma], 
                     [max(y)*0.8, mean-sigma, sigma, max(y)*0.8, mean+sigma, sigma], 
                     [max(y)*0.6, mean-sigma, sigma, max(y)*0.6, mean, sigma, max(y)*0.6, mean+sigma, sigma], 
                    ]
    bnd = [((0,200,0),(y.max()*1.15,1000, np.inf)), 
           ((0,200,0,0,200,0),(y.max()*1.15,1000, np.inf,y.max()*1.15,1000, np.inf)), 
           ((0,200,0,0,200,0,0,200,0),(y.max()*1.15,1000, np.inf,y.max()*1.15,1000, np.inf,y.max()*1.15,1000, np.inf)), 
          ]
    
    popt_list = []
    pcov_list = []
    for i in range(num_peaks):
        try:
            popt, pcov = curve_fit(fit_function[i], x, y, p0=initial_guess[i], bounds=bnd[i], maxfev=maxfev)
            popt_list.append(popt)
            pcov_list.append(pcov)
        except RuntimeError:
            maxfev=1000000
            popt, pcov = curve_fit(fit_function[i], x, y, p0=initial_guess[i], bounds=bnd[i], maxfev=maxfev)
            popt_list.append(popt)
            pcov_list.append(pcov)
    
        # A = popt[0]
        # x0 = popt[1]
        # sigma = popt[2]

    
    if plot == True:
        u = multipeak_fitting()
        u.plot_fitting(x, y, popt_list, fit_function[0])
        plt.show()
    else: pass
    
    if raw_data:
        return popt_list, pcov_list, x, y
    else:
        return popt_list, pcov_list
```
